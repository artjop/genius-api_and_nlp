```python

# Функция для создания сетевой модели текста песни
def create_song_network(text):
    G = nx.Graph()
    
    # Разбиваем текст на слова
    words = re.findall(r'\b\w+\b', text.lower())  
    
    # Создаем словарь для хранения связей между соседними словами
    word_connections = defaultdict(int)
    
    # Вычисляем связи между соседними словами
    for i in range(len(words) - 1):
        word1, word2 = words[i], words[i+1]
        word_connections[(word1, word2)] += 1
                
    # Добавляем вершины и ребра в граф
    for (word1, word2), weight in word_connections.items():
        G.add_edge(word1, word2, weight=weight)
    
    return G
```
